```python
import nltk
from sklearn.feature_extraction.text import TfidfVectorizer
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
import string
```

```python
punct_string=string.punctuation
```

```python
punct_string
```

```python
stopwords.words('hungarian')
```

```python
def text_preprocesser(text):
    words = word_tokenize(text)
    tokens = [w.lower() for w in words if w.lower() not in string.punctuation]
    stopw = stopwords.words('english')
    tokens = [token for token in tokens if token not in stopw]
    # remove words less than three letters
    tokens = [word for word in tokens if len(word)>=3]
    preprocessed_text = ' '.join(tokens)
    return preprocessed_text 
```

```python
tfidf_vectorizer = TfidfVectorizer(preprocessor=text_preprocesser)
```

```python
tfidf = tfidf_vectorizer.fit_transform(documents)
```

```python
import pandas as pd
df= pd.DataFrame(tfidf.toarray().transpose(), index=tfidf_vectorizer.get_feature_names())
```
